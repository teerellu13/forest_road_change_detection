```python
import matplotlib.pyplot as plt
import rasterio
from rasterio.plot import show
from matplotlib.cm import ScalarMappable
from matplotlib.colors import Normalize

tif = r"E:\muutostesti_leena_lokakuu2025\hakkuuluokittelu\dif3m_prop_thr_result_testi1.tif"

fig, ax = plt.subplots(figsize=(10, 8))

with rasterio.open(tif) as src:
    data = src.read(1)
    im = show(data, ax=ax, cmap="terrain")

# Create colorbar using the same cmap and data range
norm = Normalize(vmin=data.min(), vmax=data.max())
sm = ScalarMappable(norm=norm, cmap="terrain")
sm.set_array([])

cbar = fig.colorbar(sm, ax=ax)
cbar.set_label("Value")

plt.show()
```
