```python
import laspy 
import numpy as np
import os
from sklearn.neighbors import KDTree
import py4dgeo
import py4dgeo.util

import open3d as o3d



file_path_5p = r"C:\Users\telukkari\Documents\Data\nls_5p_roof_top2_half.las"
file_path_20p = r"C:\Users\telukkari\Documents\Data\nls20p_roof_top2_half.laz"

# epoch1_stable_2019, epoch2_stable_2025 = py4dgeo.read_from_las(
#     r"C:\Users\telukkari\Documents\Data\nls_5p_roof_top2_half.las", r"C:\Users\telukkari\Documents\Data\nls20p_roof_top2_half.laz"
# )

epoch1_stable_2019, epoch2_stable_2025 = py4dgeo.read_from_las(
    r"E:\ALS_DATA\registration_error_esttimate_clouds_5p_20p\nls5p_roof_top.las", r"E:\ALS_DATA\registration_error_esttimate_clouds_5p_20p\nls20p_roof_top.laz"
)

# Instantiate the M3C2 algorithm object, using all points of 2020 as corepoints (no subsampling)
m3c2 = py4dgeo.M3C2(
    epochs=(epoch1_stable_2019, epoch2_stable_2025),
    corepoints=epoch1_stable_2019.cloud[::],
    normal_radii=(1.0,),
    cyl_radii=(1.0,),
    max_distance=(5.0),
    registration_error=(0.0),
)

# Run the distance computation
m3c2_distances_stableparts, uncertainties_stableparts = m3c2.run()

# Print statistics of nn distances
print(f"Median M3C2 distances: {np.nanmedian(m3c2_distances_stableparts):.3f} m")
print(f"Mean M3C2 distances: {np.nanmean(m3c2_distances_stableparts):.3f} m")
print(f"Std. dev. of M3C2 distances: {np.nanstd(m3c2_distances_stableparts):.3f} m")

reg_2019_2025 = np.nanstd(m3c2_distances_stableparts)
print(
    f"Registration error based on point distances in stable parts is {reg_2019_2025:.3f} m."
)
```
