```python
import os
import numpy as np
import laspy
import matplotlib.pyplot as plt


folder_path = "E:\Data\laz_files_split_along_road_vectors_with_dl_predictions"

legths_of_road_patches = []

for file in os.listdir(folder_path)[:10000]:
    file_path = os.path.join(folder_path, file)
    las = laspy.read(file_path)
    number_of_5m_segments = len(np.unique(las.splitcode))
    # if number_of_5m_segments > 100:
    #     print(file)
    legths_of_road_patches.append(number_of_5m_segments)


plt.figure(figsize=(8, 5))
plt.hist(legths_of_road_patches, bins=35, edgecolor='black', alpha=0.7)

plt.xlabel("Length of patch")
plt.ylabel("Number of patches")
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
