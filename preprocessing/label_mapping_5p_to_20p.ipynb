```python
import laspy 
import numpy as np
import os
from sklearn.neighbors import KDTree

# folder_path_5p = r"D:\ALS_DATA\EVO_road_data_NLS_5p\cloud_test1_big_round_ALS.las\5m"
folder_path_5p = r"E:\ALS_DATA\NLS_5p_PADASJOKI_road_patches\laz_files_split_along_road_vectors_with_dl_predictions"
# folder_path_20p = r"D:\ALS_DATA\EVO_road_data_NLS_20p\cloud_test1_big_round_ALS\5m"
folder_path_20p = r"E:\ALS_DATA\NLS_20p_EVO_road_patches"

output_folder = r"E:\ALS_DATA\NLS_20p_EVO_road_patches_with_preds"
os.makedirs(output_folder, exist_ok=True)

for file in os.listdir(folder_path_20p):
    data_5p_laz = os.path.splitext(file)[0] + ".laz"

    # data_5p = os.path.join(folder_path_5p, file)
    data_5p = os.path.join(folder_path_5p, data_5p_laz)

    data_20p = os.path.join(folder_path_20p, file)

    points_5p = laspy.read(data_5p)
    points_20p = laspy.read(data_20p)

    xyz_5p = np.vstack((points_5p.x, points_5p.y, points_5p.z)).T
    xyz_20p = np.vstack((points_20p.x, points_20p.y, points_20p.z)).T

    predictions_5p = points_5p.prediction

    tree = KDTree(xyz_5p)
    dist, idx = tree.query(xyz_20p, k=1)

    predictions_20p = predictions_5p[idx.flatten()]

    points_20p.add_extra_dim(
        laspy.ExtraBytesParams(
            name="prediction",
            type=np.uint8, 
        )
    )
    points_20p.prediction = predictions_20p

    output_path = os.path.join(output_folder, file)
    points_20p.write(output_path)

    print(f"Saved mapped file: {output_path}")

```
